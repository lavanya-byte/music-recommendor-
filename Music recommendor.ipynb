```python
song['Singer/Artists']=song['Singer/Artists'].apply(lambda x:x.replace(" ",""))
# song['Singer/Artists']=song['Singer/Artists'].apply(lambda x:x.split(" "))
# song['Genre']=song['Genre'].apply(lambda x:x.split(" "))
song['Album/Movie']=song['Album/Movie'].apply(lambda x:x.replace(" ",""))
# song['Album/Movie']=song['Album/Movie'].apply(lambda x:x.split(" "))
```

```python
song['tag']=song['Album/Movie']+" "+song['Genre']+" "+song['Singer/Artists']
```

```python
song.head()
```

```python
new_song=song[['Song-Name','tag']]
```

```python
new_song.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count=CountVectorizer(max_features=2500)
```

```python
vector=count.fit_transform(new_song['tag']).toarray()
```

```python
vector.shape
```

```python
count.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar=cosine_similarity(vector)
#basically its calculating the distance
```

```python
# cosine_similarity(vector).shape
# total distance calculating
```

```python
similar.shape
```

```python
sorted(list(enumerate(similar[155])),reverse=True,key=lambda x:x[1])[1:6]
#ts giving the the most similar song name by sorting in reverse direction
```

```python
new_song[new_song['Song-Name']=='Aankh Marey'].index[0]
```

```python
def recommend(name):
    song_index=new_song[new_song['Song-Name']==name].index[0]
    distance=sorted(list(enumerate(similar[song_index])),reverse=True,key=lambda x:x[1])
    for i in distance[1:6]:
        print(new_song.iloc[i[0]])
```
